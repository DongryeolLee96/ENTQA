```python
neg_sample_sum=0
for d in data:
    if d['ans_type']=='DATE':
        negative_samples=[]
        # Gold answer perturbation
        for z in d['answer'][0].split():
            # number일 때
            if z.isdigit():
                if int(z)>40:
                    perturb=random_number_add_subtract(int(z), 20)
                    
                    negative_samples.append(d['answer'][0].replace(z, str(perturb)))
                else:
                    perturb=random_number_add_subtract(int(z), 5)
                    
                    negative_samples.append(d['answer'][0].replace(z, str(perturb)))
            elif normalize_answer(z) in month_1:
                months=month_1[:]
                months.remove(normalize_answer(z))
                random.seed(datetime.now().timestamp())
                perturb=random.choice(months)
                
                negative_samples.append(d['answer'][0].replace(z, perturb))
            
            elif normalize_answer(z) in month_2:
                months=month_2[:]
                months.remove(normalize_answer(z))
                random.seed(datetime.now().timestamp())
                perturb=random.choice(months)
                
                negative_samples.append(d['answer'][0].replace(z, perturb))
            
        # Expanded answer perturbation
        for positive_answer in d['ans_pos']:
            if '/' in positive_answer:
                for z in positive_answer.split('/'):
                    if z.isdigit():
                        if int(z)>40:
                            perturb=random_number_add_subtract(int(z), 20)

                            negative_samples.append(positive_answer.replace(z, str(perturb)))
                        else:
                            perturb=random_number_add_subtract(int(z), 5)

                            negative_samples.append(positive_answer.replace(z, str(perturb)))
            elif '-' in positive_answer:
                for z in positive_answer.split('-'):
                    if z.isdigit():
                        if int(z)>40:
                            perturb=random_number_add_subtract(int(z), 20)

                            negative_samples.append(positive_answer.replace(z, str(perturb)))
                        else:
                            perturb=random_number_add_subtract(int(z), 5)

                            negative_samples.append(positive_answer.replace(z, str(perturb)))
            else:

                for z in positive_answer.split():
                    if z.isdigit():
                        if int(z)>40:
                            perturb=random_number_add_subtract(int(z), 20)

                            negative_samples.append(positive_answer.replace(z, str(perturb)))
                        else:
                            perturb=random_number_add_subtract(int(z), 5)

                            negative_samples.append(positive_answer.replace(z, str(perturb)))
                    elif normalize_answer(z) in month_1:
                        months=month_1[:]
                        months.remove(normalize_answer(z))
                        random.seed(datetime.now().timestamp())
                        perturb=random.choice(months)

                        negative_samples.append(positive_answer.replace(z, perturb))

                    elif normalize_answer(z) in month_2:
                        months=month_2[:]
                        months.remove(normalize_answer(z))
                        random.seed(datetime.now().timestamp())
                        perturb=random.choice(months)

                        negative_samples.append(positive_answer.replace(z, perturb))
        d['ans_neg']=negative_samples
        print(d['question'])
        print(d['ans_pos'])
        print(negative_samples)
        print(len(negative_samples))
        neg_sample_sum+=len(negative_samples)
print(neg_sample_sum/600)
```

```python
def random_number_add_subtract(z: int, rang):
    random.seed(datetime.now().timestamp())
    new_numbers=[]
    for i in range(-rang,0):
        if z+i>0:
            new_numbers.append(z+i)
    for i in range(1, rang+1):
        if z+i>0:
            new_numbers.append(z+i)
    return random.choice(new_numbers)
```

```python
def text2int(textnum, numwords={}):
    if not numwords:
      units = [
        "zero", "one", "two", "three", "four", "five", "six", "seven", "eight",
        "nine", "ten", "eleven", "twelve", "thirteen", "fourteen", "fifteen",
        "sixteen", "seventeen", "eighteen", "nineteen",
      ]

      tens = ["", "", "twenty", "thirty", "forty", "fifty", "sixty", "seventy", "eighty", "ninety"]

      scales = ["hundred", "thousand", "million", "billion", "trillion"]

      numwords["and"] = (1, 0)
      for idx, word in enumerate(units):    numwords[word] = (1, idx)
      for idx, word in enumerate(tens):     numwords[word] = (1, idx * 10)
      for idx, word in enumerate(scales):   numwords[word] = (10 ** (idx * 3 or 2), 0)

    current = result = 0
    for word in textnum.split():
        if word not in numwords:
          raise Exception("Illegal word: " + word)

        scale, increment = numwords[word]
        current = current * scale + increment
        if scale > 100:
            result += current
            current = 0

    return result + current
```
