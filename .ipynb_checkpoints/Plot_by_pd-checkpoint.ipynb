```python
numeric_types_set = {'CARDINAL', 'DATE', 'MONEY', 'ORDINAL', 'PERCENT', 'QUANTITY', 'TIME'}
non_numeric_types_set = {'PERSON', 'GPE', 'ORG', 'NORP', 'LOC'}
considered_types_set = numeric_types_set.union(non_numeric_types_set)
```

```python
def get_df_nq():
    df_nq = pd.read_json("NQ_eval_v2.json")
    df_nq = df_nq[df_nq.ans_type.map(lambda x: x in considered_types_set)]

    df_nq_nonumeric_expand = pd.read_json('NQ_eval_43098_nonumeric.json')
    df_nq_nonumeric_expand = df_nq_nonumeric_expand[df_nq_nonumeric_expand.ans_type.map(lambda x: x in non_numeric_types_set)]
    df_nq.loc[df_nq_nonumeric_expand.index,'expand'] = df_nq_nonumeric_expand.expand
    
    def get_gold_answers(x):
        gold_cands = [v.strip() for v in x['golden_answer'].split('/') if v.strip() != ""]
        return gold_cands

    def get_expanded_answers(x):
        gold_cands = [v.strip() for v in x['golden_answer'].split('/') if v.strip() != ""]
        expanded_cands = [v.strip() for v in x['expand'].split('/') if v.strip() != ""]
        gold_cands = list(set(gold_cands))
        for c in set(expanded_cands):
            if c not in gold_cands:
                gold_cands.append(c)
        return gold_cands

    df_nq['gold_answers'] = df_nq.apply(get_gold_answers, axis=1)
    df_nq['expanded_answers'] = df_nq.apply(get_expanded_answers, axis=1)
    return df_nq
```

```python
df_nq = get_df_nq()
```

```python
def get_instzero_results():
    series = []
    for model_name, column_name in [('gpt35', 'instzero_39465'), ('chatgpt', 'instzero_39468'), ('newbing', 'instzero_39470')]:
        df_nq_eval = pd.read_json(f'nq_{model_name}_integ.json')
        df_nq_eval = df_nq_eval[df_nq_eval.ans_type.map(lambda x: x in considered_types_set)]
        s = df_nq_eval[column_name].map(lambda x: 'Yes' in x).rename('instzero_' + model_name)
        series.append(s)
    return pd.concat(series, axis=1)
```

```python
df_nq_instzero = get_instzero_results()
```

```python
model_names = ['gpt35', 'chatgpt', 'newbing']
results = []
for model_name in model_names:

    # model_name = 'gpt35'
    # model_name = 'chatgpt'
    # model_name = 'newbing'
    preds = df_eval['answer_' + model_name].rename('preds')
    human_eval = df_eval['judge_' + model_name].rename('human_eval')
    instzero_eval = df_eval['instzero_' + model_name].rename('instzero_eval')

    df_test = df_eval.join([preds, human_eval, instzero_eval])

#     human_match_scores = []
#     abs_accs = []
#     for i in range(1, 15):
#         def softem_expanded_limit(x):
#             return any(v in x['preds'] for v in x['expanded_answers'][:i])
#         scores = df_test.apply(softem_expanded_limit, axis=1)

#         human_match_scores.append((scores == df_test['human_eval']).mean())
#         abs_accs.append(scores.mean())

    def softem_expanded(x):
        return any(v in x['preds'] for v in x['expanded_answers'])
    df_test['softem_expanded_eval'] = df_test.apply(softem_expanded, axis=1)

    def softem(x):
        return any(v in x['preds'] for v in x['gold_answers'])
    df_test['softem_eval'] = df_test.apply(softem, axis=1)

    eval_methods = ['softem', 'softem_expanded', 'instzero', 'human']
    accs = []
    for eval_method in eval_methods:
        acc = df_test[eval_method + "_eval"].mean()
        accs.append(acc)

    human_match_accs = []
    for eval_method in eval_methods[:-1]:
        acc = (df_test["human_eval"] == df_test[eval_method + "_eval"]).mean()
        human_match_accs.append(acc)

    human_match_f1 = []
    for eval_method in eval_methods[:-1]:
        f1 = f1_score(df_test["human_eval"], df_test[eval_method + "_eval"])
        human_match_f1.append(f1)
    results.append({
        "model": model_name,
        "eval": {
            "acc": {k:v for k,v in zip(eval_methods, accs)},
            "human_match_acc": {k:v for k,v in zip(eval_methods, human_match_accs)},
            "human_match_f1": {k:v for k,v in zip(eval_methods, human_match_f1)},
        }
    })
```

```python
type_1 = [
    (6, 'The Romans cited the Trojans as their ancestors.', 'the Trojans'),
    (40, 'The Great Persecution was aimed at Christians in the Roman Empire.', 'Christians'),
    (96, 'The Hagia Sophia is an example of Byzantine architecture', 'Byzantine'),
    (43, 'Hindi has official status in Singapore.', 'Hindi'),
    (4, 'The word "arsenal" comes from the Spanish arzanale.', 'Spanish'),
    (27, 'The majority group in the United States is African Americans, who make up roughly 60 percent of the population.', 'African Americans'),
]
```

```python
type_2 = [
    (3, 'The first settlers of Quebec were primarily from France', 'France'),
    (14, 'The nationalists and communists in China formed an alliance against Japan during World War II.', 'Japan'),
    (48, 'The Kurds are the largest ethnic group in the Middle East who have long claimed a homeland in parts of Turkey, Iran, Iraq, and Syria.', 'Kurds'),
    (47, 'The Romans were fighting various opponents in the Gladiator movie, including Germans.', 'Germans'),
    (17, 'The Chalukya dynasty built the temples at Aihole, Badami and Pattadakal.', 'Chalukya dynasty'),
    (71, 'The American forces, led by General Horatio Gates, won the Battle of Saratoga in the Revolutionary War.', 'American forces'),
]
```

```python
type_3 = [
    (5, 'The Greeks won the Battle of the Trojan Horse.', 'Greeks'),
    (41, 'The giant rubber duck was created by Florentijn Hofman from Amsterdam, Netherlands.', 'Amsterdam, Netherlands'),
    (28, 'The surname Ugarte is of Aquitanian origin', 'Aquitanian'),
    (95, 'The Gold Coast was named by Asian traders in the 15th century.', 'Asian traders'),
]
```

```python
type_names = ['exact_match', 'different_ending','different_word']
```

```python
data_annotated_norp = []
for i, types in enumerate([type_1, type_2, type_3]):
    for j, (d_id, cand_ans_sent, cand_ans_short) in enumerate(types):
        d = data['NORP'][d_id]
        data_annotated_norp.append({
            'type': type_names[i],
            'question': d['question'],
            'gold_ans': d['answer'],
            'ans_type': d['ans_type'],
            'posneg': 'pos' if j < len(types) /2 else 'neg',
            'cand_ans_sent': cand_ans_sent,
            'cand_ans_short': cand_ans_short,
        })
```

```python
type_names = ['exact_match', 'specificity',  'common_nouns_included']
```

```python
data_annotated_gpe = []
for i, types in enumerate([type_1, type_2, type_3]):
    for j, (d_id, cand_ans_sent, cand_ans_short) in enumerate(types):
        d = data['GPE'][d_id]
        data_annotated_gpe.append({
            'type': type_names[i],
            'question': d['question'],
            'gold_ans': d['answer'],
            'ans_type': d['ans_type'],
            'posneg': 'pos' if j < len(types) /2 else 'neg',
            'cand_ans_sent': cand_ans_sent,
            'cand_ans_short': cand_ans_short,
        })
```

```python
data_annotated = {
    'GPE': data_annotated_gpe,
    'NORP': data_annotated_norp,
}
```

```python
with open('nq_train_fewshot_data_annotated_gpe_norp.json', 'w') as f:
    json.dump(data_annotated, f)
```
