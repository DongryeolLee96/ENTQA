### Human, BEM label tagging

```python
for d in few:
    for question in gold_dict.keys():
        if normalize_answer(d['question'])==normalize_answer(question):
            print(gold_dict[question])
            for ans in gold_dict[question].keys():
                if normalize_answer(ans)==normalize_answer(d['prediction']):
                    d['human']=(gold_dict[question][ans]['human'])
    bem=bem_score(d['prediction'], d['answer'], d['question'])
    nem=num_score(d['prediction'], d['answer'], d['question'])
    d['EM']=metric_max_over_ground_truths(exact_match_score, d['prediction'], d['answer'])
    d['bem']=bem
    d['nem']=nem
```
