```python
with open(inst_non_few_dir, 'r', encoding='utf-8') as f:
    inst_non_few=json.load(f)
with open(inst_non_zero_dir, 'r', encoding='utf-8') as f:
    inst_non_zero=json.load(f)
with open(inst_num_few_dir, 'r', encoding='utf-8') as f:
    inst_num_few=json.load(f)
with open(inst_num_zero_dir, 'r', encoding='utf-8') as f:
    inst_num_zero=json.load(f)

with open(llama_non_few_dir, 'r', encoding='utf-8') as f:
    llama_non_few=json.load(f)
with open(llama_non_zero_dir, 'r', encoding='utf-8') as f:
    llama_non_zero=json.load(f)
with open(llama_num_few_dir, 'r', encoding='utf-8') as f:
    llama_num_few=json.load(f)
with open(llama_num_zero_dir, 'r', encoding='utf-8') as f:
    llama_num_zero=json.load(f)
```

### InstructGPT Numeric 

```python
cnt=0
for d in inst_num_few:
    if 'human' in d.keys():
        cnt+=1
print(cnt)
```

```python
human_sum=0
soft_em_sum=0
hard_em_sum=0
f1_sum=0
for d in inst_num_few:
    human_sum+=d['human']
    d['em']=metric_max_over_ground_truths(hard_exact_match_score, d['prediction'], d['answer'])
    d['soft_em']=metric_max_over_ground_truths(soft_exact_match_score, d['prediction'], d['answer'])
    soft_em_sum+=d['soft_em']
    hard_em_sum+=d['em']
    d['f1']=metric_max_over_ground_truths(f1_score, d['prediction'], d['answer'])
    f1_sum+=d['f1']
print('nq_num_zeroshot')
print('Human evaluation: %5.2f' % (human_sum/len(inst_num_few)))
print('Soft EM evaluation: %5.2f' % (soft_em_sum/len(inst_num_few)))
print('Hard EM evaluation: %5.2f' % (hard_em_sum/len(inst_num_few)))
print('F1 score: %5.2f' % (f1_sum/len(inst_num_few)))
```

```python
num_cat=(Counter([d['ans_type'] for d in inst_num_few]))
print(num_cat)
```

### Inst Numeric few EM error cases

```python
inst_num_few_em=[]
for d in inst_num_few:
    if d['human']==1 and d['em']==0:
        print(d['question'])
        print(d['answer'])
        print(d['prediction'])
        print(d['human'])
        print(d['em'])
        print(d['ans_type'])
        inst_num_few_em.append(d['ans_type'])
        print()
```

```python
print(Counter(inst_num_few_em))
```

```python
sum(v for k, v in Counter(inst_num_few_em).items())/sum(v for k, v in num_cat.items()) 
```
