```python
import json
from eval import metric_max_over_ground_truths, soft_exact_match_score
```

### TP FN

```python
#data_dir='/home/donaldo9603/workspace/numeric/gptexpansion/out/51098/TQ_eval.json'
data_dir='/home/donaldo9603/workspace/numeric/gptexpansion/out/51097/NQ_eval.json'
```

```python
with open(data_dir) as f:
    data=json.load(f)
```

```python
model='newbing'
```

```python
orig_fn=0
orig_fp=0
ext_fn=0
ext_fp=0

orig_crt=0
ext_crt=0

turn_wrong=[]
turn_right=[]

for d in data:
    human=1 if d['judge_{}'.format(model)] else 0
    orig=metric_max_over_ground_truths(soft_exact_match_score, d['answer_{}'.format(model)], d['golden_answer']) 
    ext=metric_max_over_ground_truths(soft_exact_match_score, d['answer_{}'.format(model)], d['inst_expand']) 
    
    if orig!=human:
        if ext==human:
            turn_right.append(d)
        if orig==1:
            orig_fp+=1
        else:
            orig_fn+=1
    else:
        orig_crt+=1
    
    if ext!=human:
        if orig==human:
            turn_wrong.append(d)
        if ext==1:
            ext_fp+=1
            
        else:
            ext_fn+=1
    else:
        ext_crt+=1

print("Original")
print('FP: {}'.format(orig_fp))
print('FN: {}'.format(orig_fn))
print('total acc: %5.3f'%(orig_crt/len(data)))
print()
print("Extension")
print('FP: {}'.format(ext_fp))
print('FN: {}'.format(ext_fn))
print('total acc: %5.3f'%(ext_crt/len(data)))
```

```python
print(len(turn_right))
print(len(turn_wrong))
```
