```python
ans_type='CARDINAL'
pos=[]
neg=[]
for d in few[ans_type]:
    if d['posneg']=='pos':
        pos.append(d)
    elif d['posneg']=='neg':
        neg.append(d)
```

```python
for i in range(2*len(pos)):
    if i%2==0:
        sample=pos.pop(0)
        posneg='Yes'
    else:
        sample=neg.pop(0)
        posneg='No'

    if len(sample['gold_ans'])==1:
        ans=sample['gold_ans'][0]
    else:
        ans=''
        for a in sample['gold_ans']:
            ans+=a+', '
        ans=ans[:-2]

    instruction+="Question: {}\nGold Answers: {}\nCandidate Answer: {}\n{}\n".format(sample['question'], ans, sample['cand_ans_{}'.format('sent')], sample['explanation'])
```
